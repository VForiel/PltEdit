```python
import numpy as np
import matplotlib.pyplot as plt
import pltedit as plte

fig = plt.figure()
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.show()
plte.save(fig, "figure.plt")
```

```python
fig = plte.load("figure.plt")

# CHange matplotlib style
plt.style.use("seaborn-v0_8")

ax = fig.get_axes()[0]
ax.set_title("Loaded Figure")
ax.set_xlim(0, 5)
fig.show()
```
